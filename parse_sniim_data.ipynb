```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt, matplotlib.dates as mdates
import numpy as np
import seaborn as sns
import pprint
```

```python
path = '../data/sniim/frutas_hortalizas/'
all_files = glob.glob(os.path.join(path , "*.csv"))
len(all_files)
```

```python
def id_name_to_name(id_name):
  return id_name.replace("_"," ").capitalize().strip()

items = []
for file in all_files:
  id_name = file.split("/")[-1].split(".")[0].replace("sniim_producto_","")
  item = dict(id_name=id_name, category="Frutas y Hortalizas", type="Verdura", name=id_name_to_name(id_name))
  items.append(item)
print(len(items))
pprint.pprint(items)
```

```python
dfs = []
for file in all_files:
  df = pd.read_csv(file)
  dfs.append(df)

all_products = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
def fecha_to_date(f):
  fs = f.split('/')
  return '-'.join([fs[2],fs[1],fs[0]])

all_products['date'] = all_products['fecha'].apply(fecha_to_date)
```
